## 3) Exploratory Data Analysis (EDA) and Visualization  (18 marks) 
### Campaign Performance:  
➢ Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI. 


```python
plt.figure(figsize = (12,6))
sns.scatterplot(data = df, x ="ROI",y="Acquisition_Cost",color = ["purple"])
plt.xlabel("Acquisition_Cost",color = 'brown',fontsize = 12)
plt.ylabel("ROI",color='brown',fontsize = 12)
plt.title("Acquisition_Cost VS ROI",color = 'purple',fontsize = 12)
plt.show()
```

#### ➢ Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.
 

```python
# Average Conversion_Rate for each Channels_Used and Campaign_Type
avg_data = df.groupby(["Channel_Used", "Campaign_Type"])["Conversion_Rate"].mean().reset_index()
avg_data
```

```python
plt.figure(figsize=(14,6))
ax = sns.barplot(data=avg_data,x="Channel_Used",y="Conversion_Rate",hue="Campaign_Type",palette="crest",edgecolor='black')
    
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type="edge")

plt.title("Average Conversion Rate by Channel Used and Campaign Type", fontsize=14, weight='bold')
plt.xlabel("Channels Used",fontsize=14,color ="green",weight='bold')
plt.ylabel("Average Conversion Rate",fontsize=14,color ="green",weight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### ➢ Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot. 

```python
df.boxplot(column='Engagement_Score',by='Campaign_Type',grid=True,figsize=(10,6))
plt.title("Distribution of Engagement Score across Campaign Types",color='red',weight='bold',fontsize=14) 
plt.xlabel("Campaign Type", fontsize=14,color ="green")
plt.ylabel("Engagement_Score",fontsize=14,color ="green")
plt.suptitle("")   
plt.tight_layout()
plt.show()
```

#### ➢ Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies.


```python
avg_roi = df.groupby("Company")["ROI"].mean().reset_index()
avg_roi
```

```python
# Sort values 
avg_roi = avg_roi.sort_values(by="ROI", ascending=False)
plt.figure(figsize=(12,6))
ax = sns.barplot(data=avg_roi, x="Company", y="ROI", hue="Company",palette="Blues", edgecolor='black')
#value labels on each bar
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type="edge")
plt.title("Average ROI by Company", fontsize=14, color='purple', weight='bold')
plt.xlabel("Company", fontsize=12,color='purple', weight='bold')
plt.ylabel("Average ROI", fontsize=12,color='purple', weight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### ➢ Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap. 


```python
correlation_matrix = df[ ['Engagement_Score','Conversion_Rate']].corr()
correlation_matrix
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(correlation_matrix,annot = True, linewidths= 1)
plt.title("correlation between Engagement_Score and Conversion_Rate", fontsize=13, weight='bold')
plt.xticks(rotation= 45)
plt.show()
```

#### ➢ Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart. 


```python
df['Date'] = pd.to_datetime(df['Date'],errors='coerce')
# Creating 'YearMonth' column for grouping
df['YearMonth'] = df['Date'].dt.to_period('M')
```

```python
print(df['Date'])
```

```python
#Conversion Rate by month and company
monthly_data = df.groupby(['YearMonth', 'Company'])['Conversion_Rate'].mean().reset_index()

#Converting YearMonth back to datetime for plotting
monthly_data['YearMonth'] = monthly_data['YearMonth'].dt.to_timestamp()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_data, x='YearMonth', y='Conversion_Rate', hue='Company', marker='o', linewidth=2)

plt.title("Trend of Conversion Rate Over Time for Each Company", fontsize=14, weight='bold', color='purple')
plt.xlabel("Month", fontsize=12, color='green', weight='bold')
plt.ylabel("Conversion Rate", fontsize=12, color='green', weight='bold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### ➢ Examine the trend of Engagement_Score over Date with a line chart.

## Geographic Analysis: 
#### ➢ Determine which location has the highest Acquisition_Cost using a bar chart. 

```python
Loc_grp = df.groupby(['Location'])['Acquisition_Cost'].mean().reset_index()
# Sort by descending
Loc_grp = Loc_grp.sort_values(by='Acquisition_Cost', ascending=False)
print(Loc_grp)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=Loc_grp, x='Location', y='Acquisition_Cost', hue='Location', palette='Set1', edgecolor='black')

plt.title("Average Acquisition Cost by Location", fontsize=14, weight='bold', color='purple')
plt.xlabel("Location", fontsize=12, weight='bold', color='green')
plt.ylabel("Acquisition_Cost", fontsize=12, weight='bold', color='green')
plt.yticks(np.arange(0,6500,500))
plt.xticks(rotation=45)
plt.grid(True,alpha = 0.5)
# Highlighting Highest Bar
highest = Loc_grp.iloc[0]
plt.text(x=0, y=highest['Acquisition_Cost'] + 1, s=f"Highest: {highest['Acquisition_Cost']:.2f}", fontweight='bold', color='brown')
plt.show()
```
